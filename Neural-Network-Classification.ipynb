```python
#Creating data
from sklearn.datasets import make_circles
samples = 1000
X, y = make_circles(samples,
                    noise=0.1,
                    random_state=42)
print(X)
print(y)
```

```python
import tensorflow as tf
import numpy as np
```

```python
##Creating a basic model
tf.random.set_seed(42)

#Creating the model
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

#Compiling the model
model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.SGD(),
               metrics=["accuracy"])

#Fitting the model
model_1.fit(X,y, epochs=10)
```

```python
#training for longer
model_1.fit(X, y, epochs=100, verbose=0)
model_1.evaluate(X,y)
```

```python
#Adding more layers
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(1),
])

model_2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.SGD(),
               metrics=["accuracy"])

model_2.fit(X, y, epochs=100, verbose=0)
model_2.evaluate(X,y)
```

```python
## IMPROVING THE MODEL

#add more layers or increase the no. of hidden units
model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),
])

#different optimizer, different learning rates
model_3.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["accuracy"])

#More epochs
model_3.fit(X, y, epochs=100, verbose=0)

#Evaluating the model
model_3.evaluate(X, y)
```

```python
#Plotting the decision boundary
def plot_decision_boundary(model, X, y):
    #Plots the decision boundary created by a model predicting on X
    
    #Defining axis boundaries
    x_min, x_max = X[:,0].min()- 0.1, X[:,0].max()+ 0.1
    y_min, y_max = X[:,1].min()- 0.1, X[:,1].max()+ 0.1
    
    ##creating a meshgrid
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                        np.linspace(y_min, y_max, 100))
    
    #creating X value
    x_in = np.c_[xx.ravel(), yy.ravel()] #Stacks 2D arrays together
    
    #Predicting values
    y_pred = model.predict(x_in)
    
    #Checking for multi-class classification
    if(len(y_pred[0])>1):
        print("Multiclass Classification")
        #Reshaping
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Binary Classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
    
    #Plotting the decision boundary
    plt.contourf(xx,yy,y_pred, cmap=plt.cm.RdYlBu, alpha=0.6)
    plt.scatter(X[:,0],X[:,1],c=y, s=30, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
#Checking the predictions of the model
plot_decision_boundary(model_3, X, y)
```

```python
#Using the model for a regression problem
X_regression = tf.range(0,1000,5)
y_regression = tf.range(100,1100,5)

#splitting the data
X_reg_train = X_regression[:150]
X_reg_test = X_regression[150:]

y_reg_train = y_regression[:150]
y_reg_test = y_regression[150:]

#Changing the loss function and fitting the model
model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),
])
model_4.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["mae"]) #BinaryCrossentropy will not work with regression problem
model_4.fit(tf.expand_dims(X_reg_train,axis=-1), y_reg_train, epochs=100)
```

```python
y_reg_pred = model_4.predict(X_reg_test)

plt.figure(figsize=(10,10))
plt.scatter(X_reg_train,y_reg_train, c="b", label="Training data")
plt.scatter(X_reg_test, y_reg_test, c="g", label="Testing data")
plt.scatter(X_reg_test, y_reg_pred, c="r", label="Prediction data")
plt.legend()
```

```python
#Creating the tensopr
A = tf.cast(tf.range(-10,10), tf.float32)
A
```

```python
#visualising the tensor
plt.plot(A)
```

```python
#Writing the sigmoid function from scratch
def sigmoid(x):
    return 1/(1+tf.exp(-x))
```

```python
#Applying activation function directly to tensor
B = sigmoid(A)
B
```

```python
plt.plot(B)
```

```python
#Writing relu from scratch
def relu(x):
    return tf.maximum(x,0)
```

```python
C = relu(A)
C
```
