```python
#Plotting the decision boundary
def plot_decision_boundary(model, X, y):
    #Plots the decision boundary created by a model predicting on X
    
    #Defining axis boundaries
    x_min, x_max = X[:,0].min()- 0.1, X[:,0].max()+ 0.1
    y_min, y_max = X[:,1].min()- 0.1, X[:,1].max()+ 0.1
    
    ##creating a meshgrid
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                        np.linspace(y_min, y_max, 100))
    
    #creating X value
    x_in = np.c_[xx.ravel(), yy.ravel()] #Stacks 2D arrays together
    
    #Predicting values
    y_pred = model.predict(x_in)
    
    #Checking for multi-class classification
    if(len(y_pred[0])>1):
        print("Multiclass Classification")
        #Reshaping
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Binary Classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
    
    #Plotting the decision boundary
    plt.contourf(xx,yy,y_pred, cmap=plt.cm.RdYlBu, alpha=0.6)
    plt.scatter(X[:,0],X[:,1],c=y, s=30, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
#Checking the predictions of the model
plot_decision_boundary(model_3, X, y)
```

```python
#Callback to find the ideal learning rate

tf.random.set_seed(42)

model_10 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model_10.compile(loss="binary_crossentropy",
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

#callback function
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

history_10 = model_10.fit(X_train, y_train, epochs=100, callbacks=[lr_scheduler])
```

```python
##MULTICLASS CLASSIFICATION

#Importing the dataset
from tensorflow.keras.datasets import fashion_mnist #This has already been split into training and testing data
(train_data, train_label), (test_data, test_label) = fashion_mnist.load_data()

train_data.shape, test_data.shape, train_label.shape
```

```python
#Creating a list of all the classes
classes = ["T-shirt","Trouser","Pullover", "Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
len(classes), classes
```

```python
#plotting an example image
plt.imshow(train_data[7])
classes[train_label[7]]
```

```python
#Building a model
model_multi_1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)), #converts 2D array into 1D
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax"),
])

model_multi_1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(), #Sparse has to be used when labels are not one-hot encodedb
                     optimizer=tf.keras.optimizers.Adam(),
                     metrics=["accuracy"])

history_multi_1 = model_multi_1.fit(train_data, train_label, epochs=5,
                                   validation_data = (test_data, test_label))
```

```python
#model summary
model_multi_1.summary()
```

```python
#Improving performance with normalisation - Image pixel values to be between 0 and 1

#Normalising the data
train_data_norm = train_data/255
test_data_norm = test_data/255
```

```python
#Fitting the model
history_multi_1_norm = model_multi_1.fit(train_data_norm, train_label,
                                   epochs=5,
                                   validation_data=(test_data_norm,test_label))
```

```python
#Comparing normalised and non-normalised data
pd.DataFrame(history_multi_1.history).plot(title="Non-Normalised")

pd.DataFrame(history_multi_1_norm.history).plot(title="Normalised")
```

```python
#As it can be seen loss curves are lower and accuracy curves are higher
```

```python
#Finding ideal learning rate
model_multi_2 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model_multi_2.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                     optimizer=tf.keras.optimizers.Adam(),
                     metrics=["accuracy"])

#callback for finding learning rate
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 *10**(epoch/20))

histiry_multi_2 = model_multi_2.fit(train_data_norm, train_label,
                                   epochs=40,
                                   validation_data=(test_data_norm, test_label),
                                   callbacks=[lr_scheduler])
```

```python
#plotting the learning rate decay curve
lrs = 1e-3 *10**((np.arange(40))/20)
plt.semilogx(lrs, histiry_multi_2.history['loss'])
plt.xlabel("learning rate")
plt.ylabel("loss")
```

```python
#Ideal learning rate is 1e-3 - default learning rate of Adam optimizer

#refitting the model with this learning rate
model_multi_2.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                     optimizer=tf.keras.optimizers.Adam(0.001),
                     metrics=["accuracy"])
history_multi_2_opt = model_multi_2.fit(train_data_norm, train_label,
                                       validation_data=(test_data_norm, test_label),
                                       epochs=20)
```

```python
#Evaluating the model
y_pred = model_multi_2.predict(test_data_norm)
y_pred = np.argmax(y_pred, axis=1)
```
